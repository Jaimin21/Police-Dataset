# Q.5  Compare the age distributions for each violation.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))

# Create a box plot to compare age distributions for each violation
sns.boxplot(x='violation', y='driver_age', data=data)

# Add title and labels
plt.title('Age Distributions for Each Violation')
plt.xlabel('Violation')
plt.ylabel('Driver Age')

# Display the plot
plt.show()
```
